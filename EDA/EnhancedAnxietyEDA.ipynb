# columns explanation

- Age – How old are you?
- Gender – Are you male or female?
- Occupation – What do you do? (Job, student, homemaker, etc.)
- Sleep Hours – How many hours do you sleep daily?
- Physical Activity (hrs/week) – How many hours of exercise or activity do you do in a week?
- Caffeine Intake (mg/day) – How much caffeine do you have daily? (like tea, coffee, energy drinks)
- Alcohol Consumption (drinks/week) – How many alcoholic drinks do you have in a week?
- Smoking – Do you smoke or vape?
- Family History of Anxiety – Does anyone in your family have anxiety issues?
- Stress Level (1-10) – How stressed are you? (1 = super chill, 10 = super stressed)
- Heart Rate (bpm) – Your heartbeat per minute.
- Breathing Rate (breaths/min) – How many breaths you take in a minute.
- Sweating Level (1-5) – How much do you sweat? (1 = almost none, 5 = a lot)
- Dizziness – Do you feel dizzy or lightheaded?
- Medication – Are you taking any medicines?
- Therapy Sessions (per month) – How many therapy/counseling sessions do you have in a month?
- Recent Major Life Event – Any big change or incident recently? (e.g. breakup, job change, loss)
- Diet Quality (1-10) – How healthy is your diet? (1 = mostly junk, 10 = very healthy)
- Anxiety Level (1-10) – How anxious do you feel? (1 = calm, 10 = full panic)

```python
copy_df = df.copy()
```

```python
copy_df.rename(columns={
    'Age': 'age',
    'Gender': 'gender',
    'Occupation': 'occupation',
    'Sleep Hours': 'sleep_hours',
    'Physical Activity (hrs/week)': 'physical_activity',
    'Caffeine Intake (mg/day)': 'caffeine_intake',
    'Alcohol Consumption (drinks/week)': 'alcohol_consumption',
    'Smoking': 'smoking',
    'Family History of Anxiety': 'family_anxiety_history',
    'Stress Level (1-10)': 'stress_level',
    'Heart Rate (bpm)': 'heart_rate',
    'Breathing Rate (breaths/min)': 'breathing_rate',
    'Sweating Level (1-5)': 'sweating_level',
    'Dizziness': 'dizziness',
    'Medication': 'medication',
    'Therapy Sessions (per month)': 'therapy_sessions',
    'Recent Major Life Event': 'major_life_event',
    'Diet Quality (1-10)': 'diet_quality',
    'Anxiety Level (1-10)': 'anxiety_level'
}, inplace=True)
```

- Anxiety & Stress: Anxiety shows the strongest correlation with stress levels (0.67), indicating high stress often accompanies anxiety.
- Sleep & Anxiety: More sleep hours correlate with lower anxiety (-0.49), highlighting sleep's protective role.
- Therapy & Anxiety: Higher therapy sessions link to higher anxiety (0.52), suggesting therapy is sought by those with elevated anxiety.
- Caffeine & Anxiety: Increased caffeine intake correlates with higher anxiety (0.35), aligning with caffeine's stimulant effects.
- Diet & Anxiety: Better diet quality associates with lower anxiety (-0.22), emphasizing nutrition's mental health impact.

```python
corr = copy_df.select_dtypes(include='number').corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(copy_df)
plt.suptitle("Relationships Between Key Variables")
plt.show()
```

- Gender Differences in Anxiety: The graph likely shows variations in anxiety levels among females, males, and other genders, highlighting potential disparities in mental health experiences.
- Possible Trends: If data were visible, we might see if one gender reports higher anxiety levels, which could inform targeted mental health support.

```python
sns.barplot(data=copy_df, x='gender', y='anxiety_level' , ci=None)
plt.title('Anxiety Level by Gender')
plt.show()
```

- Positive Correlation Likely: The graph suggests that as stress levels (X-axis) increase, anxiety levels (Y-axis) also tend to rise, forming an upward trend. This aligns with the well-established link between stress and anxiety.

```python
sns.regplot(data=copy_df, x='stress_level', y='anxiety_level')
plt.title('Stress vs Anxiety Level')
plt.show()
```

- Higher caffeine intake correlates with elevated anxiety levels, suggesting that excessive consumption may worsen anxiety symptoms.
- Moderate or low caffeine intake shows minimal anxiety impact, highlighting a potential threshold for tolerance.

```python
sns.barplot(data=copy_df, x='caffeine_intake', y='anxiety_level',ci = None )
plt.title('Caffeine Intake vs Anxiety Level')

plt.show()
```

- Non-Linear Relationship: Stress levels initially dip slightly (from 6.1 to 5.2) with low alcohol consumption (0–4 units), suggesting temporary relief.
- Rising Stress at Higher Consumption: Beyond ~5 units, stress levels climb again, indicating alcohol may worsen stress with excessive use.

```python
ax = sns.barplot(data=copy_df, x='alcohol_consumption', y='stress_level' ,ci = None)
plt.title('Alcohol Consumption vs Stress Level')

for bar in ax.patches:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.2, f'{height:.1f}', ha='center')

plt.show()
```

- Smokers Show Higher Avg. Breathing Rate (50.5%) compared to non-smokers (49.5%), suggesting that smoking may slightly elevate respiratory activity—likely due to lung irritation or reduced oxygen efficiency.

```python
# Group and calculate average breathing rate per smoking group
avg_breathing = copy_df.groupby('smoking')['breathing_rate'].mean()

plt.figure(figsize=(6, 6))
plt.pie(
    avg_breathing,
    labels=[f'{label} ({value:.1f}%)' for label, value in zip(avg_breathing.index, avg_breathing / avg_breathing.sum() * 100)],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('pastel')
)
plt.title('Share of Avg. Breathing Rate by Smoking Status')
plt.axis('equal')
plt.show()
```

- Family History = Higher Anxiety: Individuals with a family history of anxiety (55.2%) report significantly higher anxiety levels (avg. ~5/6) compared to those without (44.8%, avg. ~2/6).
- Genetic/Environmental Link: The 10.4% gap suggests family history (genes or upbringing) plays a notable role in anxiety susceptibility.

```python
ax = sns.barplot(data=copy_df, x='family_anxiety_history', y='anxiety_level' , ci = None)
plt.title('Anxiety Level by Family History')

total = copy_df.groupby('family_anxiety_history')['anxiety_level'].mean().sum()

for bar in ax.patches:
    pct = bar.get_height() / total * 100
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, f'{pct:.1f}%', ha='center')

plt.ylim(0, ax.get_ylim()[1] + 2)
plt.show()
```

- Positive Correlation Likely: The graph suggests that as anxiety levels rise (X-axis), heart rate (Y-axis) also tends to increase—consistent with the body's stress response .

```python
sns.barplot(data=copy_df, x='heart_rate', y='anxiety_level' , ci = None)
plt.title('Heart Rate vs Anxiety Level')
plt.xticks(rotation=100) 
plt.show()
```

```python
sns.barplot(data=copy_df, x='age', y='diet_quality',ci = None)
plt.title('Age vs Diet Quality')
plt.xticks(rotation = 100)
plt.show()
```

- Therapy Reduces Anxiety: Higher therapy session numbers (e.g., 8–12) correlate with lower anxiety levels (0–2), suggesting effective symptom management.

```python
sns.scatterplot(data=copy_df, x='therapy_sessions', y='anxiety_level')
plt.title('Therapy Sessions vs Anxiety Level')
plt.show()
```

- Major Life Events Spike Stress/Anxiety: Individuals reporting a recent major life event show 30.5% higher stress/anxiety levels compared to those without (19.2%).
- Baseline Stress Exists: Even without major events, ~20% of people experience stress/anxiety, highlighting other contributing factors.

```python
# Reshape the data to long format
df_melted = copy_df.melt(id_vars='major_life_event',
                         value_vars=['stress_level', 'anxiety_level'],
                         var_name='Measure', value_name='Score')

# Barplot
ax = sns.barplot(data=df_melted, x='major_life_event', y='Score', hue='Measure',ci = None)
plt.title('Stress and Anxiety by Recent Major Life Event')

total = df_melted.groupby(['major_life_event', 'Measure'])['Score'].mean().sum()
for bar in ax.patches:
    height = bar.get_height()
    pct = height / total * 100
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.2, f'{pct:.1f}%', ha='center')

plt.ylim(0, ax.get_ylim()[1] + 2)
plt.show()
```

- Highest Stress:

Healthcare roles (Nurse, Doctor) and Lawyers show the highest stress (5-6/6), likely due to high-stakes workloads and emotional demands.
Artists/Musicians also rank high (~4-5/6), possibly from financial instability or creative pressures.

- Lower Stress:

Engineers and Teachers report moderate stress (3-4/6), balancing structure with challenges.
Students and Freelancers vary widely, reflecting situational factors (e.g., exams, income volatility).



```python
sns.barplot(data=copy_df, x='occupation', y='stress_level' , ci = None)
plt.title('Occupation vs Stress Level')
plt.xticks(rotation=45)
plt.show()
```

- The highest stress/anxiety scores correlate with sweating levels 4 and 5 (12.5% each), suggesting moderate sweating may signal peak stress.
- Lower sweating (levels 1–3) shows reduced stress (7–11.2%), while higher levels (6–9) taper off, possibly indicating exhaustion or adaptation.

```python
df_melted = copy_df.melt(id_vars='sweating_level',
                         value_vars=['stress_level', 'anxiety_level'],
                         var_name='Measure', value_name='Score')

# Create the barplot
ax = sns.barplot(data=df_melted, x='sweating_level', y='Score', hue='Measure',ci = None)
plt.title('Stress and Anxiety by Sweating Level')
plt.xlabel('Sweating Level')
plt.ylabel('Average Score')

# Add percentage labels
total = df_melted.groupby(['sweating_level', 'Measure'])['Score'].mean().sum()
for bar in ax.patches:
    height = bar.get_height()
    pct = height / total * 100
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.2, f'{pct:.1f}%', ha='center')
                                            
plt.ylim(0, ax.get_ylim()[1] + 2)
plt.show()
```

- Dizziness Linked to Higher Anxiety: 52.5% of individuals reporting dizziness experience elevated anxiety levels, compared to 47.5% without dizziness.
- Actionable Insight: Dizziness could be a physical marker for anxiety screening, especially in clinical settings.

```python
ax = sns.barplot(data=copy_df, x='dizziness', y='anxiety_level' , ci = None)
plt.title('Anxiety Level by Dizziness')

total = copy_df.groupby('dizziness')['anxiety_level'].mean().sum()

for bar in ax.patches:
    height = bar.get_height()
    pct = height / total * 100
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{pct:.1f}%', ha='center')

plt.ylim(0, ax.get_ylim()[1] + 1)
plt.show()
```

- Positive Correlation: As physical activity increases (X-axis), sleep hours (Y-axis) also rise, peaking at ~8 hours for high activity levels (8–10).
- Optimal Range: Moderate activity (4–6) aligns with 6–8 hours of sleep, suggesting balanced exercise improves sleep quality.

```python
sns.scatterplot(data=copy_df, x='physical_activity', y='sleep_hours')
plt.title('Physical Activity vs Sleep Hours')
plt.show()
```

- Higher physical activity levels (8–10) correlate with lower anxiety levels (2–4), while sedentary behavior (0–2) links to higher anxiety (6–8).

```python
sns.regplot(data=copy_df, x='physical_activity', y='anxiety_level')
plt.title('Physical Activity vs Anxiety Level')
plt.show()
```

# Insights

- Physical Activity Reduces Anxiety: Higher activity levels consistently correlate with lower anxiety, showing exercise's mental health benefits.

- Sedentary Lifestyle Risks: Low activity (0–3) links to high anxiety (6+), highlighting inactivity as a risk factor.

- Moderate Exercise is Key: Anxiety drops sharply beyond activity level 4, suggesting even 30+ minutes daily can help.

- Non-Linear Relationship: Anxiety declines steeply at first, then plateaus, implying diminishing returns at very high activity levels.

- Actionable Insight: Aim for at least moderate activity (4–6/10) to significantly lower anxiety symptoms.
