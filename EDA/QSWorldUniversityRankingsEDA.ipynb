# Columns Explanations

1. 2026 Rank – The university’s rank in the year 2026.
2. Previous Rank – The rank of the university in the previous year (2025).
3. Institution Name – Name of the university or college.
4. Country/Territory – The country where the university is located.
5. Region – The region or continent (like Asia, Europe) the university belongs to.
6. Size – The size of the university (number of students, etc.).
7. Focus – The university’s main focus (like teaching, research, etc.).
8. Research – How active the university is in research work.
9. Status – Whether the university is public or private.
10. AR SCORE – Score based on academic reputation (what experts/professors think).
11. AR RANK – Rank based on academic reputation score.
12. ER SCORE – Score based on how employers value the university’s graduates.
13. ER RANK – Rank based on employer reputation.
14. FSR SCORE – Score based on number of students per faculty member.
15. FSR RANK – Rank based on that ratio.
16. CPF SCORE – Score based on how often faculty research is cited by others.
17. CPF RANK – Rank based on citation score.
18. IFR SCORE – Score based on how many foreign/international faculty members are there.
19. IFR RANK – Rank based on that.
20. ISR SCORE – Score based on how many international students are there.
21. ISR RANK – Rank based on that.
22. ISD SCORE – Score for global research collaborations.
23. ISD RANK – Rank based on that score.
24. IRN SCORE – Score showing quality of international research.
25. IRN RANK – Rank based on that.
26. EO SCORE – Score based on how successful students are in getting jobs.
27. EO RANK – Rank based on employment outcome.
28. SUS SCORE – Score based on how environmentally sustainable the university is.
29. SUS RANK – Rank based on that.
30. verall SCORE – Final overall score combining all the above metrics.

# Rename all columns

```python
df.rename(columns={
    '2026 Rank': 'Rank_2026',
    'Previous Rank': 'Rank_Prev',
    'Institution Name': 'Institution',
    'Country/Territory': 'Country',
    'Region': 'Region',
    'Size': 'Size',
    'Focus': 'Focus',
    'Research': 'Research',
    'Status': 'Status',
    'AR SCORE': 'AR_Score',
    'AR RANK': 'AR_Rank',
    'ER SCORE': 'ER_Score',
    'ER RANK': 'ER_Rank',
    'FSR SCORE': 'FSR_Score',
    'FSR RANK': 'FSR_Rank',
    'CPF SCORE': 'CPF_Score',
    'CPF RANK': 'CPF_Rank',
    'IFR SCORE': 'IFR_Score',
    'IFR RANK': 'IFR_Rank',
    'ISR SCORE': 'ISR_Score',
    'ISR RANK': 'ISR_Rank',
    'ISD SCORE': 'ISD_Score',
    'ISD RANK': 'ISD_Rank',
    'IRN SCORE': 'IRN_Score',
    'IRN RANK': 'IRN_Rank',
    'EO SCORE': 'EO_Score',
    'EO RANK': 'EO_Rank',
    'SUS SCORE': 'SUS_Score',
    'SUS RANK': 'SUS_Rank',
    'Overall SCORE': 'Overall_Score'
}, inplace=True)
```

- USA has the highest number of universities in the QS 2026 rankings.
- Other top contributors include the UK, China, and India.

```python
# Which country's universities get more ranks (count barplot)
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Country', order=df['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

- MIT ranks #1 globally, followed by Imperial, Stanford, and Oxford.
- These top universities are from USA, UK, and a few from Asia-Pacific.

```python
# which university has highest rank 
df['Rank_2026'] = pd.to_numeric(df['Rank_2026'], errors='coerce')
top20 = df.nsmallest(20, 'Rank_2026')

plt.figure(figsize=(10,6))
sns.barplot(data=top20, y='Institution', x='Rank_2026', palette='viridis')
plt.title("Top 20 Universities by 2026 Rank")
plt.tight_layout()
plt.show()
```

- IIT Delhi and IIT Bombay are India's best ranked universities.
- Total 54 Indian universities are in the dataset, These 5 are in top 220.

```python
# Total Indian universities & which are in top 5 overall

# Get all Indian universities
indian_unis = df[df['Country'] == 'India']

# Get top 5 Indian universities 
indian_top5 = indian_unis.nsmallest(5, 'Rank_2026')

plt.figure(figsize=(10, 7))
sns.barplot(data=indian_top5, 
            x='Institution', 
            y='Rank_2026',
            palette='viridis')

plt.title(f"Top 5 Indian Universities (Total: {len(indian_unis)})")
plt.xlabel("University Name")
plt.ylabel("2026 Rank")
plt.xticks(rotation=45)
plt.tight_layout()  # Prevents label cutoff
plt.show()

```

- Smaller universities tend to have better (higher) FSR scores.
- As size increases, FSR score decreases — possibly due to crowding.

```python
#Larger universities tend to have lower Faculty-Student Ratio
sns.barplot(data=df, x='Size', y='FSR_Score' , ci = None)
plt.title("Faculty-Student Ratio by University Size")
plt.xlabel("Size")
plt.ylabel("FSR Score")
plt.show()
```

- European universities score higher in sustainability compared to Asian ones.
- This may reflect stronger environmental practices and reporting in Europe.

```python
#Universities in Europe have better Sustainability scores than Asia
region_df = df[df['Region'].isin(['Asia', 'Europe'])]
sns.barplot(data=region_df, x='Region', y='SUS_Score' , ci = None)
plt.title("Sustainability Scores: Europe vs Asia")
plt.xlabel("Region")
plt.ylabel("Sustainability Score")
plt.show()
```

- Developed countries have a much higher International Faculty Ratio.
- This shows better faculty diversity and global hiring practices.

```python
# Developed countries have more international faculty
development_map = {
    'India': 'Developing', 'USA': 'Developed', 'UK': 'Developed', 'Germany': 'Developed',
    'Australia': 'Developed', 'Japan': 'Developed', 'China': 'Developing',
    'Canada': 'Developed', 'France': 'Developed', 'South Korea': 'Developed'
}
df['Development'] = df['Country'].map(development_map)

sns.barplot(data=df, x='Development', y='IFR_Score' , ci = None)
plt.title("International Faculty Ratio: Developed vs Developing")
plt.xlabel("Development Status")
plt.ylabel("IFR Score")
plt.show()
```

- Higher AR Scores strongly correlate with higher Overall Scores, showing reputation directly impacts rankings.
- The near-linear trend suggests AR Score is a key component of overall evaluation.

```python
#Academic Reputation vs Overall Score (correlation)

df['AR_Score'] = pd.to_numeric(df['AR_Score'], errors='coerce')
df['Overall_Score'] = pd.to_numeric(df['Overall_Score'], errors='coerce')

sns.regplot(data=df, 
            x='AR_Score', 
            y='Overall_Score', 
            scatter_kws={'alpha':0.6},
            line_kws={'color':'red'})
plt.title("Academic Reputation vs Overall Score")
plt.xlabel("AR Score")
plt.ylabel("Overall Score")
plt.show()
```

- Private not-for-profit universities have the highest employer reputation.
- Private-for-profit universities rank lowest in ER scores.

```python
# Private vs Public universities: Employer Reputation
sns.barplot(data=df, x='Status', y='ER_Score' , color = 'yellow' , errorbar=None)
plt.title("Employer Reputation by University Type")
plt.xlabel("University Type")
plt.ylabel("ER Score")
plt.show()
```

- Universities with more international students (high ISR) tend to have stronger global research networks.
- The positive relationship highlights how diversity boosts research collaboration.

```python
#Intl Student Ratio vs Intl Research Network
sns.regplot(data=df, x='ISR_Score', y='ISD_Score', scatter_kws={'alpha':0.6})
plt.title("Intl Student Ratio vs Intl Research Network")
plt.xlabel("ISR Score")
plt.ylabel("ISD Score")
plt.show()

```

- Higher international faculty percentages correlate with better academic reputation scores.
- Global faculty recruitment appears to enhance institutional prestige.

```python
# Intl Faculty Ratio vs Academic Reputation
sns.regplot(data=df, x='IFR_Score', y='AR_Score', scatter_kws={'alpha':0.6})
plt.title("Intl Faculty Ratio vs Academic Reputation")
plt.xlabel("IFR Score")
plt.ylabel("AR Score")
plt.show()
```

- "VH" (Very High) research category institutions show the highest citation rates (80-100 CPF).
- Research quality tiers clearly separate by citation performance.

```python
#Citations per Faculty vs Research
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Research_Encoded'] = le.fit_transform(df['Research'].astype(str))

plt.figure(figsize=(10,6))
sns.regplot(
    data=df.dropna(subset=['CPF_Score', 'Research_Encoded']),
    x='CPF_Score',
    y='Research_Encoded',
    scatter_kws={'alpha':0.6, 'color':'blue'},
    line_kws={'color':'red', 'linewidth':2}
)

# Customize y-axis to show original labels
unique_labels = df['Research'].unique()
plt.yticks( ticks=range(len(unique_labels)), labels=unique_labels)

plt.title("Citations per Faculty vs Research Category")
plt.xlabel("CPF Score")
plt.ylabel("Research Category")
plt.tight_layout()
plt.show()
```

- Stronger research networks correspond to higher overall scores, especially above 60 ISD.
- International collaboration is crucial for top-ranked institutions.

```python
# Intl Research Network vs Overall Score
sns.regplot(data=df, x='ISD_Score', y='Overall_Score', scatter_kws={'alpha':0.6})
plt.title("Intl Research Network vs Overall Score")
plt.xlabel("ISD Score")
plt.ylabel("Overall Score")
plt.show()

```

- Better employer reputations (ER) align with superior graduate employment results (EQ).
- The tight clustering suggests reputation accurately predicts job success.

```python
# Employer Reputation vs Employment Outcome
sns.regplot(data=df, x='ER_Score', y='EO_Score', scatter_kws={'alpha':0.6})
plt.title("Employer Reputation vs Employment Outcome")
plt.xlabel("ER Score")
plt.ylabel("EO Score")
plt.show()
```

# Insights

- English-speaking & developed countries dominate top ranks.

Academic Reputation,
Employer Reputation,
Overall Score.
